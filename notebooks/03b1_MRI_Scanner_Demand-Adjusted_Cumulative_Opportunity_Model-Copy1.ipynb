# Endoscopy Demand & Accessibility Mapping â€“ Enhanced Version (2024)

This notebook replicates and improves upon the earlier endoscopy spatial demand model by integrating the following enhancements:

Robust site access logic: ensuring no LSOA directly on a facility is marked as underserved.

Cleaner pipeline: streamlining imports, 
GeoDataFrame merges, and cumulative accessibility logic.

Travel matrix validation: checks for intra-LSOA facility access and missing travel entries.

Integrated overlays: consistent facility marker placement (CDC and Acute) with choropleths.

Improved export handling: clear file paths and geometry separation.

```python
# Standard imports
import os
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
import seaborn as sns
import contextily as ctx
import warnings

# Optional display settings
warnings.filterwarnings("ignore")
pd.set_option("display.max_columns", 100)

# ----------------------------------------------------------
# File Path Setup (MRI)
# ----------------------------------------------------------

# Base directory structure
base_dir = "/Users/rosstaylor/Downloads/Research Project/Code Folder/diagnostic-modality-demand/diagnostic-modality-demand"
data_dir = os.path.join(base_dir, "data")
processed_dir = os.path.join(data_dir, "processed")
raw_dir = os.path.join(data_dir, "raw")

# Explicit modality input folder
modality_dir = os.path.join(raw_dir, "modality_modelling")

# Input file paths (MRI)
lsoa_gpkg_path = os.path.join(processed_dir, "LSOA_MRI_Demand_with_Demographics_2024_v1.gpkg")
cdc_path = os.path.join(raw_dir, "health_infra", "NHS_SW_ Community_Diagnostic_Centres_enriched.csv")  # Note extra space
acute_path = os.path.join(modality_dir, "NHS_SW_Acute_Hospitals_Capability.csv")
travel_matrix_path = os.path.join(raw_dir, "LSOA_to_LSOA_complete.csv")

# Output folder
output_dir = os.path.join(base_dir, "outputs")
os.makedirs(output_dir, exist_ok=True)
```

```python
# Load spatial demand file (GeoPackage with LSOA geometry and MRI demand)
try:
    gdf = gpd.read_file(lsoa_gpkg_path, layer="mri_demand")
except Exception:
    # Fallback: read the first layer if 'mri_demand' isn't present
    gdf = gpd.read_file(lsoa_gpkg_path)
print(f"LSOA MRI GPKG loaded with shape: {gdf.shape}")

# Load site capability files
df_cdc = pd.read_csv(cdc_path)
df_acute = pd.read_csv(acute_path)
print(f"Loaded CDC capability data: {df_cdc.shape[0]} rows")
print(f"Loaded Acute hospital data: {df_acute.shape[0]} rows")

# Load LSOA-to-LSOA travel time matrix
df_travel = pd.read_csv(travel_matrix_path)
print(f"Travel matrix loaded with shape: {df_travel.shape[0]} rows")

# Quick previews
display(gdf.head(2))
display(df_cdc.head(1))
display(df_acute.head(1))
display(df_travel.head(2))
```

```python
# ----------------------------------------------------------
# Step 1a â€“ Concatenate CDC and Acute to form df_sites (MRI)
# ----------------------------------------------------------

# Add consistent 'Type' column if missing (required for later filtering)
df_cdc["Type"] = "CDC"
df_acute["Type"] = "Acute"

# Concatenate both sources
df_sites = pd.concat([df_cdc, df_acute], axis=0, ignore_index=True)

# Filter: only sites with >0 MRI scanners and valid LSOA codes
df_sites = df_sites[df_sites["MRI Scanners"].fillna(0) > 0].copy()
df_sites = df_sites[df_sites["lsoa21cd"].notna()].copy()

# Standardise LSOA code type
df_sites["lsoa21cd"] = df_sites["lsoa21cd"].astype(str)
```

```python
# ----------------------------------------------------------
# Step 2 â€“ Compute Scanner-Weighted Accessibility (MRI)
# ----------------------------------------------------------

# 2.1 â€“ Create scanner-weighted supply per destination LSOA (MRI)
scanner_supply = (
    df_sites[["lsoa21cd", "MRI Scanners"]]
    .groupby("lsoa21cd", as_index=False)
    .sum()
    .rename(columns={"lsoa21cd": "destination_name", "MRI Scanners": "mri_scanner_supply"})
)

# 2.2 â€“ Patch travel matrix with missing intra-LSOA self-links (0 minutes travel time)
# First, get all LSOAs from the population demand file
try:
    df_lsoa = gpd.read_file(lsoa_gpkg_path, layer="mri_demand")[["lsoa21cd"]]  # prefer MRI layer
except Exception:
    df_lsoa = gpd.read_file(lsoa_gpkg_path)[["lsoa21cd"]]  # fallback if layer name differs
df_lsoa["lsoa21cd"] = df_lsoa["lsoa21cd"].astype(str)

all_lsoa_codes = set(df_lsoa["lsoa21cd"])
existing_self_links = set(
    df_travel.query("origin_name == destination_name")["origin_name"]
)
missing_self_links = all_lsoa_codes - existing_self_links

# Construct missing rows with travel_time_car/PT = 0 for self-travel
patch_df = pd.DataFrame({
    "origin_name": list(missing_self_links),
    "destination_name": list(missing_self_links),
    "travel_time_car": 0,
    "travel_time_pt_peak": 0,
    "travel_time_pt_nonpeak": 0
})

# 2.3 â€“ Concatenate with original travel matrix
df_travel_patched = pd.concat([df_travel, patch_df], ignore_index=True)

# 2.4 â€“ Merge in MRI scanner supply per destination
df_travel_supply = df_travel_patched.merge(
    scanner_supply,
    on="destination_name",
    how="left"
)

# Optional: Fill NaN with 0 where scanner supply was missing
df_travel_supply["mri_scanner_supply"] = df_travel_supply["mri_scanner_supply"].fillna(0)

# Preview
print(f"Patched travel matrix shape: {df_travel_supply.shape}")
display(df_travel_supply.head())
```

```python
# ----------------------------------------------------------
# Step 3a â€“ Prepare CDC and Acute GeoDataFrames (MRI)
# (robust to missing gdf_lsoa)
# ----------------------------------------------------------

# Ensure we have an LSOA geometry frame with ['lsoa21cd','geometry']
if 'gdf_lsoa' not in globals() or gdf_lsoa is None:
    if 'gdf' in globals() and {'lsoa21cd','geometry'}.issubset(set(gdf.columns)):
        gdf_lsoa = gdf[['lsoa21cd','geometry']].copy()
        # keep CRS if present
        if hasattr(gdf, 'crs'):
            gdf_lsoa = gpd.GeoDataFrame(gdf_lsoa, geometry='geometry', crs=gdf.crs)
        print("[INFO] Built gdf_lsoa from existing 'gdf'.")
    else:
        try:
            gdf_lsoa = gpd.read_file(lsoa_gpkg_path, layer="mri_demand")[["lsoa21cd", "geometry"]].copy()
            print("[INFO] Built gdf_lsoa from MRI GPKG (layer='mri_demand').")
        except Exception:
            gdf_lsoa = gpd.read_file(lsoa_gpkg_path)[["lsoa21cd", "geometry"]].copy()
            print("[INFO] Built gdf_lsoa from MRI GPKG (first layer).")

# Assign site type from existing 'Type' column (df_sites should already be MRI-filtered)
df_sites["site_type"] = df_sites["Type"]

# Merge site LSOAs with geometry from gdf_lsoa
facilities_gdf = df_sites.merge(gdf_lsoa[["lsoa21cd", "geometry"]], on="lsoa21cd", how="left")
facilities_gdf = gpd.GeoDataFrame(facilities_gdf, geometry="geometry", crs=gdf_lsoa.crs)

# Use centroids of LSOA polygons for plotting facility locations
facilities_gdf["geometry"] = facilities_gdf["geometry"].centroid

# Create separate GeoDataFrames for CDC and Acute sites
cdc_gdf = facilities_gdf[facilities_gdf["site_type"] == "CDC"].copy()
acute_gdf = facilities_gdf[facilities_gdf["site_type"] == "Acute"].copy()

print(f"[OK] facilities_gdf: {facilities_gdf.shape}, CDC: {cdc_gdf.shape}, Acute: {acute_gdf.shape}")
```

```python
# ----------------------------------------------------------
# Step 3 â€“ Compute Scanner-Weighted Accessibility Scores (MRI)
# ----------------------------------------------------------

# 3.1 â€“ Create time band flags (car mode)
df_travel_supply["lt_20"]  = df_travel_supply["travel_time_car"] < 20
df_travel_supply["20_40"]  = df_travel_supply["travel_time_car"].between(20, 40, inclusive="left")
df_travel_supply["gt_40"]  = df_travel_supply["travel_time_car"] > 40

# 3.2 â€“ Initialise accessibility DataFrame
access_df = (
    df_travel_supply[["origin_name"]]
    .drop_duplicates()
    .set_index("origin_name")
)

# 3.3 â€“ Group and sum MRI scanner supply for each band
lt20    = df_travel_supply[df_travel_supply["lt_20"]].groupby("origin_name")["mri_scanner_supply"].sum()
btw2040 = df_travel_supply[df_travel_supply["20_40"]].groupby("origin_name")["mri_scanner_supply"].sum()
gt40    = df_travel_supply[df_travel_supply["gt_40"]].groupby("origin_name")["mri_scanner_supply"].sum()

# 3.4 â€“ Join and rename (MRI)
access_df["mri_access_car_lt_20"]  = lt20
access_df["mri_access_car_20_40"]  = btw2040
access_df["mri_access_car_gt_40"]  = gt40
access_df = access_df.fillna(0).reset_index().rename(columns={"origin_name": "lsoa21cd"})

# 3.5 â€“ Merge with LSOA base geometry (prefer MRI layer)
try:
    gdf_geom = gpd.read_file(lsoa_gpkg_path, layer="mri_demand")[["lsoa21cd", "geometry"]]
except Exception:
    gdf_geom = gpd.read_file(lsoa_gpkg_path)[["lsoa21cd", "geometry"]]

gdf_lsoa = gdf_geom.merge(access_df, on="lsoa21cd", how="left")

# 3.6 â€“ Fill unmatched LSOAs with 0
gdf_lsoa[["mri_access_car_lt_20", "mri_access_car_20_40", "mri_access_car_gt_40"]] = (
    gdf_lsoa[["mri_access_car_lt_20", "mri_access_car_20_40", "mri_access_car_gt_40"]].fillna(0)
)
```

```python
# Preview updated gdf_lsoa (MRI)
print("Sample of LSOA Accessibility Scores (MRI):")
display(gdf_lsoa[["lsoa21cd", "mri_access_car_lt_20", "mri_access_car_20_40", "mri_access_car_gt_40"]].head())

# Plot histograms for each time band (MRI)
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

gdf_lsoa["mri_access_car_lt_20"].plot.hist(bins=20, ax=axes[0], color="green", edgecolor="black")
axes[0].set_title("Access < 20 min")
axes[0].set_xlabel("MRI Scanner Supply")

gdf_lsoa["mri_access_car_20_40"].plot.hist(bins=20, ax=axes[1], color="orange", edgecolor="black")
axes[1].set_title("Access 20â€“40 min")
axes[1].set_xlabel("MRI Scanner Supply")

gdf_lsoa["mri_access_car_gt_40"].plot.hist(bins=20, ax=axes[2], color="red", edgecolor="black")
axes[2].set_title("Access > 40 min")
axes[2].set_xlabel("MRI Scanner Supply")

plt.suptitle("Distribution of MRI Scanner Accessibility by Car Travel Time", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------------------
# Step 4 â€“ Multi-Panel Map: MRI Access by Car (3 Travel Time Bands)
# ----------------------------------------------------------

import matplotlib.pyplot as plt

# Define columns and labels (MRI)
access_cols = {
    'mri_access_car_lt_20': '< 20 min',
    'mri_access_car_20_40': '20â€“40 min',
    'mri_access_car_gt_40': '> 40 min'
}

# Create subplots for each access band
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(22, 8))

for ax, (col, label) in zip(axes, access_cols.items()):
    
    # Extract geometry and access column
    gdf_band = gdf_lsoa[["lsoa21cd", "geometry", col]].copy()
    
    # Plot base LSOA layer
    gdf_band.plot(
        column=col,
        cmap="OrRd",
        linewidth=0.1,
        edgecolor="grey",
        legend=True,
        legend_kwds={"label": f"MRI Scanners Reachable ({label})", "shrink": 0.5},
        ax=ax
    )
    
    # Plot CDCs (orange)
    cdc_gdf.plot(
        ax=ax,
        color="orange",
        markersize=30,
        marker="o",
        edgecolor="white",
        linewidth=0.5,
        label="CDC"
    )
    
    # Plot Acute Hospitals (purple)
    acute_gdf.plot(
        ax=ax,
        color="purple",
        markersize=30,
        marker="^",
        edgecolor="white",
        linewidth=0.5,
        label="Acute Hospital"
    )
    
    # Format
    ax.set_title(f"MRI Access by Car â€“ {label}", fontsize=13)
    ax.axis("off")
    ax.legend(loc="lower left", frameon=True)

    # Basemap
    ctx.add_basemap(ax, crs=gdf_lsoa.crs.to_string(), source=ctx.providers.CartoDB.Positron)

# Final layout
plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------------------
# Step 5 â€“ Highlight LSOAs with No MRI Site Access in >40 mins
# ----------------------------------------------------------

# 5.1 â€“ Filter LSOAs with zero MRI access even beyond 40 minutes
gdf_zero_gt40 = gdf_lsoa[gdf_lsoa['mri_access_car_gt_40'] == 0].copy()
print(f"Number of LSOAs with 0 MRI access even at >40 mins: {gdf_zero_gt40.shape[0]}")

# 5.2 â€“ Plot map with highlighted LSOA polygons
fig, ax = plt.subplots(figsize=(10, 10))

# Base: all LSOA outlines for context
gdf_lsoa.boundary.plot(ax=ax, color='lightgrey', linewidth=0.2)

# Highlight: underserved LSOAs as filled red polygons
gdf_zero_gt40.plot(
    ax=ax,
    facecolor='red',
    edgecolor='black',
    linewidth=0.3,
    alpha=0.6,
    label='LSOA >40min No Access'
)

# Overlay: CDCs (orange circles)
cdc_gdf.plot(
    ax=ax,
    color='orange',
    marker='o',
    markersize=35,
    edgecolor='white',
    linewidth=0.5,
    label='CDC'
)

# Overlay: Acute Hospitals (purple triangles)
acute_gdf.plot(
    ax=ax,
    color='purple',
    marker='^',
    markersize=35,
    edgecolor='white',
    linewidth=0.5,
    label='Acute Hospital'
)

# Final map formatting
ax.set_title("LSOAs with No MRI Access within >40 Minutes\n(Highlighted Polygons with Facility Overlay)", fontsize=14)
ax.axis('off')
ax.legend(loc='lower left', frameon=True)

# Basemap
ctx.add_basemap(ax, crs=gdf_lsoa.crs.to_string(), source=ctx.providers.CartoDB.Positron)

plt.tight_layout()
plt.show()
```

```python
# ----------------------------------------------------------
# Step 6 â€“ Export Final LSOA Accessibility GeoDataFrame (MRI)
# ----------------------------------------------------------

# Reload full demographic and demand data from MRI GPKG
try:
    gdf_base = gpd.read_file(lsoa_gpkg_path, layer="mri_demand")[[
        "lsoa21cd", "mri_total_demand", "total_population"
    ]]
except Exception:
    # Fallback if layer name differs
    gdf_base = gpd.read_file(lsoa_gpkg_path)[[
        "lsoa21cd", "mri_total_demand", "total_population"
    ]]

# Merge with accessibility scores and geometry
gdf_lsoa_export = gdf_lsoa.merge(gdf_base, on="lsoa21cd", how="left")

# Define output path (MRI)
export_path = os.path.join(processed_dir, "gdf_lsoa_MRI_accessibility_2024.gpkg")

# Save final export (write to a clear layer name for downstream reads)
gdf_lsoa_export.to_file(export_path, layer="lsoa_access", driver="GPKG")
print(f"Export complete: {export_path}")
```
